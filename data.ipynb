# **Collect average spherical refraction and axial length of myopic children from each year**
1.   Collect the data of children whose spherical refraction(SR) < 0 from each year
2.   Calculate the average for SR and axial length from each year

# **Collecting SR_right from 2014**

```python
df[df.examination_date == 2014]
thnA = df[df.examination_date == 2014]

thnA
```

```python
thnA = df[df.examination_date == 2014]
thnA[thnA['SR_right'] < 0]
```

```python
meanAR = thnA[thnA['SR_right'] < 0]

print(meanAR.SR_right.mean())

meanAR
```

```python
ASRR14 = meanAR.SR_right.mean()
print (ASRR14)
```

# **Graph for myopia**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Line for Right Eye
x = [2014, 2015, 2016, 2017]
y = [ASRR14,ASRR15,ASRR16,ASRR17]

plt.plot(x,y, label = 'Right Eye', color='red', marker='.')
plt.title("Average Myopia Degree 2014 - 2017")
plt.xlabel("Year", fontdict={'fontsize': 10,}, fontweight="bold")
plt.ylabel("Diopters (D)", fontdict={'fontsize': 10}, fontweight="bold")

## Line for Left Eye
x2 = [2014,2015,2016,2017]
y2 = [ASRL14,ASRL15,ASRL16,ASRL17]

plt.plot(x2,y2, label = 'Left Eye', color= 'blue', marker='.')

plt.xticks([2014,2015,2016,2017])
plt.yticks([-0.56,-0.8,-0.88,-1.14,-1.25,-1.57,-1.63])

## To invert the yaxis
plt.gca().invert_yaxis()

plt.legend()

plt.show()

# To save the graph
#plt.savefig('/content/drive/MyDrive/mataData/myopia_graph.png', dpi=200, bbox_inches = 'tight')
```
